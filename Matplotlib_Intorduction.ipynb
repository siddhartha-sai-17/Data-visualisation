```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data.csv")
df = df.dropna(subset=['Duration', 'Pulse'])

# Find min and max of Duration
min_duration = df['Duration'].min()
max_duration = df['Duration'].max()

# Define bins dynamically
bins = [min_duration - 1, 20, 40, 60, 80, max_duration + 1]

# Make sure bins are strictly increasing
bins = sorted(bins)

labels = ['0-20','20-40','40-60','60-80','80+']

df.loc[:, 'Duration_group'] = pd.cut(df['Duration'], bins=bins, labels=labels, right=False)

# Compute average pulse per group
avg_pulse = df.groupby('Duration_group', observed=False)['Pulse'].mean().reset_index()

# Draw bar chart
plt.figure(figsize=(8,6))
plt.bar(avg_pulse['Duration_group'], avg_pulse['Pulse'], color='skyblue', edgecolor='black')
plt.title("Average Pulse by Duration Group")
plt.xlabel("Duration Group (minutes)")
plt.ylabel("Average Pulse")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("bar_avg_pulse_by_duration.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import pearsonr

# Load dataset
df = pd.read_csv("data.csv")

# Drop rows with missing Duration or Calories
df = df.dropna(subset=['Duration', 'Calories'])

# Compute Pearson correlation coefficient
corr_coeff, _ = pearsonr(df['Duration'], df['Calories'])

# Scatter plot
plt.scatter(df['Duration'], df['Calories'], color='orange')

# Add linear trend line
slope, intercept = np.polyfit(df['Duration'], df['Calories'], 1)
plt.plot(df['Duration'], slope*df['Duration'] + intercept, color='blue', linestyle='--')

# Labels and title with correlation
plt.xlabel("Duration")
plt.ylabel("Calories")
plt.title(f"Duration vs Calories (Pearson r = {corr_coeff:.2f})")
plt.grid(True)

# Save figure
plt.tight_layout()
plt.savefig("scatter_duration_vs_calories.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data.csv")

# Drop rows with missing Maxpulse
df = df.dropna(subset=['Maxpulse'])

# Define Maxpulse bins and labels
bins = [0, 120, 140, 160, df['Maxpulse'].max()+1]
labels = ['<=120', '121-140', '141-160', '>160']

# Categorize Maxpulse into bins
df['Maxpulse_range'] = pd.cut(df['Maxpulse'], bins=bins, labels=labels, right=True, include_lowest=True)

# Count number of entries in each range
counts = df['Maxpulse_range'].value_counts().sort_index()

# Plot pie chart
plt.figure(figsize=(5,5))
plt.pie(counts, labels=counts.index, autopct='%1.1f%%', colors=['red','darkgreen','darkblue','violet'], startangle=90)
plt.title("Maxpulse Range Breakdown")
plt.savefig("pie_maxpulse_ranges.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("data.csv")

# Drop rows with missing Calories
df = df.dropna(subset=['Calories'])

# Determine number of bins using Sturges' rule
num_bins = int(1 + 3.322 * (len(df['Calories']).bit_length()))  # approximate using log2(n)

# Plot histogram
plt.figure(figsize=(8,6))
plt.hist(df['Calories'], bins=num_bins, color='orange', edgecolor='black')
plt.title("Distribution of Calories")
plt.xlabel("Calories")
plt.ylabel("Count")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.savefig("hist_calories.png")
plt.show()

# Comment on shape/skewness
cal_mean = df['Calories'].mean()
cal_median = df['Calories'].median()
if cal_mean > cal_median:
    skewness = "positively skewed (right-skewed)"
elif cal_mean < cal_median:
    skewness = "negatively skewed (left-skewed)"
else:
    skewness = "approximately symmetric"
print(f"The histogram of Calories is {skewness}.")
```
