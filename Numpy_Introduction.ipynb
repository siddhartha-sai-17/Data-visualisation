Indexing and Slicing:
You are working as a Data Analyst at a fruit distribution company.

The company stores the number of fruit boxes shipped per day (for 1 week) in a 2D NumPy array where:
Each row represents a type of fruit.
Each column represents a day of the week (Mon–Sun).


Your Tasks
Create the 2D NumPy array

Create an array named shipments using the above data.
Access single elements (Indexing):
Get the number of Apples shipped on Wednesday.
Get the number of Oranges shipped on Sunday.
Access multiple elements (Slicing):
Extract all the weekday shipments (Mon–Fri) for all fruits.
Extract only the weekend shipments (Sat & Sun).
Extract the shipments of Apples from Wednesday to Saturday.
Bonus (Practice thinking):
Replace the Saturday shipment of Apples (currently 80) with 100 using indexing.


> Add blockquote



```python
import numpy as np

# Rows → Apples, Oranges
# Columns → Mon, Tue, Wed, Thu, Fri, Sat, Sun
shipments = np.array([
    [50, 60, 55, 70, 65, 80, 90],  # Apples
    [40, 45, 50, 60, 55, 70, 75]   # Oranges
])

print("Shipments array:\n", shipments)
```

```python
wed_apple = shipments[0,2]
print(wed_apple)
```

```python
sun_orange = shipments[1,6]
print(sun_orange)
```

```python
weekday_shipments = shipments[:, 0:5]
print("\nWeekday shipments (Mon–Fri):\n", weekday_shipments)
weekend_shipments = shipments[:, 5:7]
print("\nWeekend shipments (Sat & Sun):\n", weekend_shipments)
apples_wed_to_sat = shipments[0, 2:6]
print("\nApples (Wed–Sat):", apples_wed_to_sat)
```

```python
shipments[0,5] = 100
print("updated shipments",shipments)
```

```python
import numpy as np

# -----------------------------
# 1️⃣ Create 2D arrays for each farm (4 panels × 3 hours)
# Sample data (kWh)
farmA = np.array([[10, 12, 11],
                  [9, 14, 13],
                  [15, 16, 14],
                  [11, 12, 13]])

farmB = np.array([[8, 10, 9],
                  [12, 13, 14],
                  [10, 11, 12],
                  [9, 8, 10]])

farmC = np.array([[14, 15, 16],
                  [13, 12, 14],
                  [10, 11, 12],
                  [15, 16, 17]])

# 2️⃣ Stack into 3D array (farms × panels × hours)
production_all = np.stack([farmA, farmB, farmC], axis=0)

# 3️⃣ Array attributes
print("Shape:", production_all.shape)
print("Size:", production_all.size)
print("Dtype:", production_all.dtype)
print("Dimensions:", production_all.ndim)

# -----------------------------
# 4️⃣ Per-farm totals and stats
total_per_farm = production_all.sum(axis=(1,2))  # sum over panels & hours
mean_per_farm = production_all.mean(axis=(1,2))
std_per_farm = production_all.std(axis=(1,2))
var_per_farm = production_all.var(axis=(1,2))

print("Total per farm:", total_per_farm)
print("Mean per farm:", mean_per_farm)
print("Std per farm:", std_per_farm)
print("Var per farm:", var_per_farm)

# -----------------------------
# 5️⃣ Weighted per-panel energy (dot product with hour_weights)
hour_weights = np.array([0.5, 1.0, 0.75])
weighted_per_panel = np.dot(production_all, hour_weights)  # shape (3,4)
print("Weighted per panel:\n", weighted_per_panel)

# -----------------------------
# 6️⃣ Apply broadcasting adjustment (+0.5 kWh for Farm B)
production_all[1] += 0.5  # broadcasting across all panels & hours in Farm B

# -----------------------------
# 7️⃣ Z-score normalization per farm
# reshape mean & std for broadcasting (3,1,1)
mean_reshaped = mean_per_farm[:, np.newaxis, np.newaxis]
std_reshaped = std_per_farm[:, np.newaxis, np.newaxis]

normalized = (production_all - mean_reshaped) / std_reshaped
print("Normalized shape:", normalized.shape)

# -----------------------------
# 8️⃣ Element-wise ufuncs

# Exponential family
exp_data = np.exp(normalized)
expm1_data = np.expm1(normalized)
# np.log requires positive values, normalize to positive domain
log_data = np.log(exp_data)  # safe because exp_data > 0
log10_data = np.log10(exp_data)
log2_data = np.log2(exp_data)

# Trigonometric
sin_data = np.sin(normalized)
cos_data = np.cos(normalized)
tan_data = np.tan(normalized)
arcsin_data = np.arcsin(np.clip(normalized, -1, 1))  # clip to [-1,1]
arccos_data = np.arccos(np.clip(normalized, -1, 1))
arctan_data = np.arctan(normalized)
degrees_data = np.degrees(normalized)
radians_data = np.radians(normalized)

# Rounding
round_data = np.round(normalized, 2)
floor_data = np.floor(normalized)
ceil_data = np.ceil(normalized)
trunc_data = np.trunc(normalized)

# -----------------------------
# 9️⃣ Comparison & logical
high_output = production_all > 20
even_hours_mask = (np.arange(production_all.shape[2]) % 2 == 0)
# broadcast even_hours_mask to panels
even_hours_mask_broadcast = even_hours_mask[np.newaxis, np.newaxis, :]
logical_and_mask = np.logical_and(high_output, even_hours_mask_broadcast)
logical_or_mask = np.logical_or(high_output, even_hours_mask_broadcast)
logical_not_mask = np.logical_not(high_output)
logical_xor_mask = np.logical_xor(high_output, even_hours_mask_broadcast)

# -----------------------------
# 10️⃣ Bitwise operations (integer array)
alerts = np.array([[1,0,1],
                   [0,0,1],
                   [1,1,0],
                   [0,1,0]])
# replicate for each farm
alerts_farms = np.stack([alerts]*3, axis=0)
bitwise_and = np.bitwise_and(alerts_farms, 1)
bitwise_or = np.bitwise_or(alerts_farms, 0)
bitwise_invert = np.invert(alerts_farms.astype(np.int8))  # careful with negatives

# -----------------------------
# 11️⃣ Arithmetic ufuncs again
add_data = np.add(normalized, 2)
sub_data = np.subtract(normalized, 1)
mul_data = np.multiply(normalized, 3)
div_data = np.divide(normalized, 2)
power_data = np.power(normalized, 2)
mod_data = np.mod(normalized, 2)

# -----------------------------
# 12️⃣ Aggregations on ufunc outputs
sum_exp = np.sum(exp_data)
mean_sin = np.mean(sin_data)

# -----------------------------
# 13️⃣ Panel totals & maintenance
panel_totals_per_farm = production_all.sum(axis=2)  # sum across hours → shape (3,4)
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=1)
print("Panel totals per farm:\n", panel_totals_per_farm)
print("Needs maintenance:", needs_maintenance)

# -----------------------------
# 14️⃣ Final KPIs
overall_total = production_all.sum()
peak_hour_value = production_all.max()
peak_idx = np.unravel_index(np.argmax(production_all), production_all.shape)  # farm, panel, hour
average_per_panel = production_all.mean(axis=2)  # shape (3,4)

print("Overall total energy:", overall_total)
print("Peak hour value:", peak_hour_value, "at farm, panel, hour:", peak_idx)
print("Average per panel:\n", average_per_panel)
```

```python
import numpy as np

# -----------------------------
# 1️⃣ Create 2D arrays for each farm (panels × hours)
farmA = np.array([[10, 12, 11],
                  [9, 14, 13],
                  [15, 16, 14],
                  [11, 12, 13]])

farmB = np.array([[8, 10, 9],
                  [12, 13, 14],
                  [10, 11, 12],
                  [9, 8, 10]])

farmC = np.array([[14, 15, 16],
                  [13, 12, 14],
                  [10, 11, 12],
                  [15, 16, 17]])

# -----------------------------
# 2️⃣ Stack into 3D array (farms × panels × hours) and convert to float
production_all = np.stack([farmA, farmB, farmC], axis=0).astype(float)

# -----------------------------
# 3️⃣ Array attributes
print("Shape:", production_all.shape)
print("Size:", production_all.size)
print("Dtype:", production_all.dtype)
print("Dimensions:", production_all.ndim)

# -----------------------------
# 4️⃣ Per-farm totals and stats
total_per_farm = production_all.sum(axis=(1,2))
mean_per_farm = production_all.mean(axis=(1,2))
std_per_farm = production_all.std(axis=(1,2))
var_per_farm = production_all.var(axis=(1,2))

print("Total per farm:", total_per_farm)
print("Mean per farm:", mean_per_farm)
print("Std per farm:", std_per_farm)
print("Var per farm:", var_per_farm)

# -----------------------------
# 5️⃣ Weighted per-panel energy
hour_weights = np.array([0.5, 1.0, 0.75])  # shape (3,)
weighted_per_panel = np.dot(production_all, hour_weights)  # shape (3,4)
print("Weighted per panel:\n", weighted_per_panel)

# -----------------------------
# 6️⃣ Apply broadcasting adjustment (+0.5 kWh for Farm B)
production_all[1] += 0.5  # now works because dtype is float

# -----------------------------
# 7️⃣ Z-score normalization per farm
mean_reshaped = mean_per_farm[:, np.newaxis, np.newaxis]  # shape (3,1,1)
std_reshaped = std_per_farm[:, np.newaxis, np.newaxis]    # shape (3,1,1)

normalized = (production_all - mean_reshaped) / std_reshaped
print("Normalized shape:", normalized.shape)

# -----------------------------
# 8️⃣ Example element-wise ufuncs
exp_data = np.exp(normalized)
sin_data = np.sin(normalized)
round_data = np.round(normalized, 2)

# -----------------------------
# 9️⃣ Comparison & logical
high_output = production_all > 20
even_hours_mask = (np.arange(production_all.shape[2]) % 2 == 0)
even_hours_mask_broadcast = even_hours_mask[np.newaxis, np.newaxis, :]
logical_and_mask = np.logical_and(high_output, even_hours_mask_broadcast)

# -----------------------------
# 10️⃣ Aggregations
panel_totals_per_farm = production_all.sum(axis=2)  # sum across hours → shape (3,4)
needs_maintenance = np.any(panel_totals_per_farm < 40, axis=1)

overall_total = production_all.sum()
peak_hour_value = production_all.max()
peak_idx = np.unravel_index(np.argmax(production_all), production_all.shape)
average_per_panel = production_all.mean(axis=2)

# -----------------------------
# ✅ Print summary
print("\nPanel totals per farm:\n", panel_totals_per_farm)
print("Needs maintenance:", needs_maintenance)
print("Overall total energy:", overall_total)
print("Peak hour value:", peak_hour_value, "at farm, panel, hour:", peak_idx)
print("Average per panel:\n", average_per_panel)
```
