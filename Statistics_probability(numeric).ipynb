```python
import pandas as pd

def classify_columns(df):
    column_info = {}

    for col in df.columns:
        if pd.api.types.is_numeric_dtype(df[col]):
            unique_vals = df[col].nunique()
            if pd.api.types.is_float_dtype(df[col]) or unique_vals > 20:
                column_info[col] = "Continuous (Numeric)"
            else:
                column_info[col] = "Discrete (Numeric)"
        else:
            column_info[col] = "Discrete (Categorical)"

    return column_info


# Example
column_types = classify_columns(df)
for col, ctype in column_types.items():
    print(f"{col} → {ctype}")
```

```python
import numpy as np

def remove_outliers_zscore(df, cols):
    for col in cols:
        mean = df[col].mean()
        std = df[col].std()
        z = (df[col] - mean) / std
        df = df[np.abs(z) < 3]
    return df
```
