```python
df = pd.read_csv('Customer_support_data.csv')
```

```python
df.shape
```

```python
df = df.dropna()
```

```python
df
```

```python
sns.histplot(df['Item_price'],kde=True)
plt.title("Item price Distribution")
plt.show()
```

```python

sns.boxplot(y='CSAT Score', data=df, color='skyblue')
plt.title("CSAT Score Variation Overall")
plt.ylabel("CSAT Score")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

```python
sns.violinplot(y='connected_handling_time',data=df,color='blue')
plt.title("Spread of Connected handling time values")
plt.grid(axis='y',linestyle='--',alpha=0.7)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Count how many times each score appears
csat_counts = df['CSAT Score'].value_counts().sort_index()

# Plot pie chart
plt.figure(figsize=(6,6))
plt.pie(csat_counts, labels=csat_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("CSAT Score Distribution")
plt.show()
```
