```python
# Notebook: Walmart Store Sales - Full Pipeline (Tasks 1-8)
# Requirements: pandas, numpy, matplotlib, seaborn, scipy
# If running in Colab, uncomment the install lines below:
# !pip install pandas numpy matplotlib seaborn scipy

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from functools import reduce

sns.set(style="whitegrid")
pd.options.display.max_columns = 200
pd.options.display.width = 160

# -----------------------------
# Helper: download from GitHub raw
# -----------------------------
def load_csv_from_github(user, repo, path, filename):
    """
    Attempts to load file from https://raw.githubusercontent.com/{user}/{repo}/master/{path}/{filename}
    Returns DataFrame or raises error.
    """
    raw_url = f"https://raw.githubusercontent.com/{user}/{repo}/master/{path}/{filename}"
    print(f"Trying to load: {raw_url}")
    try:
        df = pd.read_csv(raw_url)
        print(f"Loaded {filename} (shape {df.shape})")
        return df
    except Exception as e:
        raise RuntimeError(f"Could not load {filename} from GitHub: {e}")

# -----------------------------
# Task 1 — Data Loading & Merging
# -----------------------------
# Adjust these if your file names differ.
user = "ezgigm"
repo = "Project4_Store_Sales_Forecasting"
path = "Raw%20Data"   # URL-encoded space 'Raw Data'

# Typical filenames in the Walmart dataset/project:
filenames = {
    "stores": "stores.csv",
    "train": "train.csv",
    "features": "features.csv"
}

# Try load from GitHub raw (fallback: you can load local files instead)
try:
    stores_df = load_csv_from_github(user, repo, path, filenames["stores"])
    train_df = load_csv_from_github(user, repo, path, filenames["train"])
    features_df = load_csv_from_github(user, repo, path, filenames["features"])
except Exception as e:
    print("Automatic GitHub load failed. You can instead upload files or provide local paths.")
    raise

# Quick peek at column names
print("\nstores columns:", stores_df.columns.tolist())
print("train columns:", train_df.columns.tolist())
print("features columns:", features_df.columns.tolist())

# The project expects merges on ['Store','Dept','Date'] across different files.
# Typical merge pipeline (train + features) on ['Store','Date'], then join stores on 'Store'
# 1) Merge train with features on ['Store', 'Date']
# 2) Merge result with stores on ['Store']

# Normalize column names: strip spaces and unify case
def clean_cols(df):
    df = df.copy()
    df.columns = df.columns.str.strip()
    return df

stores_df = clean_cols(stores_df)
train_df = clean_cols(train_df)
features_df = clean_cols(features_df)

# Ensure Date is present in features and train
print("\nColumn presence check:")
print("In train:", set(['Store','Dept','Date']).issubset(set(train_df.columns)))
print("In features:", set(['Store','Date']).issubset(set(features_df.columns)))
print("In stores:", 'Store' in stores_df.columns)

# Merge 1: train + features on ['Store','Date']
merged_tf = pd.merge(train_df, features_df, on=['Store', 'Date'], how='left', validate="m:1")
print("After merging train+features:", merged_tf.shape)

# Merge 2: merged_tf + stores on Store
merged_all = pd.merge(merged_tf, stores_df, on='Store', how='left', validate="m:1")
print("After merging with stores:", merged_all.shape)

# Display first and last 10 rows
display(merged_all.head(10))
display(merged_all.tail(10))

# Print shape, info, describe
print("\nShape:", merged_all.shape)
print("\nInfo():")
print(merged_all.info())

print("\nDescribe (numeric):")
display(merged_all.describe())

# -----------------------------
# Identify column types (Task 1 subitems)
# -----------------------------
# Numerical columns
numeric_cols = merged_all.select_dtypes(include=['number']).columns.tolist()
# Categorical columns (object, category)
categorical_cols = merged_all.select_dtypes(include=['object', 'category']).columns.tolist()
# Date columns — check parseable 'Date' or datetime dtype
date_cols = [c for c in merged_all.columns if 'date' in c.lower()]

print("\nNumerical columns:", numeric_cols)
print("Categorical columns:", categorical_cols)
print("Date-like columns:", date_cols)

# List all unique store types and departments (if present)
if 'Type' in merged_all.columns:
    print("\nUnique Store Types:", merged_all['Type'].unique())
else:
    print("\nColumn 'Type' not found in merged dataset.")

if 'Dept' in merged_all.columns:
    print("Unique Departments (sample 50):", merged_all['Dept'].unique()[:50])
else:
    print("Column 'Dept' not found in merged dataset.")

# Identify columns that may require cleaning or type conversion
possible_issues = []
# e.g., object columns that contain numbers, date strings, or many nulls
for col in merged_all.columns:
    if merged_all[col].dtype == 'object':
        nunique = merged_all[col].nunique(dropna=True)
        nulls = merged_all[col].isnull().sum()
        if nunique < 50 or nulls>0:
            possible_issues.append((col, merged_all[col].dtype, nunique, nulls))
print("\nColumns that may need cleaning / type conversion (sample):")
for row in possible_issues:
    print(row)

# -----------------------------
# Task 2 — Data Cleaning
# -----------------------------
df = merged_all.copy()

# 1) Identify missing values
print("\nMissing values (before cleaning):")
display(df.isnull().sum())

# 2) Fill numeric values (Temperature, Fuel_Price, CPI, Unemployment) using median
numeric_fill_median = ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']
for col in numeric_fill_median:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')   # ensure numeric
        median_val = df[col].median(skipna=True)
        df[col].fillna(median_val, inplace=True)
        print(f"Filled {col} missing with median: {median_val}")

# 3) Fill markdown-related fields with mean (if exist)
markdown_cols = [f'MarkDown{i}' for i in range(1,6)]
for col in markdown_cols:
    if col in df.columns:
        df[col] = pd.to_numeric(df[col], errors='coerce')
        mean_val = df[col].mean(skipna=True)
        df[col].fillna(mean_val, inplace=True)
        print(f"Filled {col} missing with mean: {mean_val}")

# 4) Convert Date column to datetime
if 'Date' in df.columns:
    df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
    print("Converted 'Date' to datetime. Nulls after conversion:", df['Date'].isnull().sum())

# 5) Remove duplicate rows
before_dups = df.shape[0]
df.drop_duplicates(inplace=True)
after_dups = df.shape[0]
print(f"Removed duplicates: {before_dups - after_dups}")

# 6) Reset index
df.reset_index(drop=True, inplace=True)

print("\nMissing values (after cleaning):")
display(df.isnull().sum())

# -----------------------------
# Task 3 — Outlier Detection & Treatment
# -----------------------------
# We'll provide functions to detect outliers (boxplot + IQR)
def iqr_outlier_bounds(series, k=1.5):
    q1 = series.quantile(0.25)
    q3 = series.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    return lower, upper

outlier_columns = ['Weekly_Sales', 'Temperature', 'Fuel_Price', 'CPI']
for col in outlier_columns:
    if col in df.columns:
        s = pd.to_numeric(df[col], errors='coerce').dropna()
        lower, upper = iqr_outlier_bounds(s)
        num_outliers = ((s < lower) | (s > upper)).sum()
        print(f"{col}: lower={lower:.3f}, upper={upper:.3f}, outliers={num_outliers}")
        # Boxplot
        plt.figure(figsize=(8,2))
        sns.boxplot(x=s)
        plt.title(f"Boxplot of {col}")
        plt.show()

# Identify extreme spikes during holiday weeks
if 'IsHoliday' in df.columns and 'Weekly_Sales' in df.columns:
    holiday_sales = df[df['IsHoliday'] == True]['Weekly_Sales']
    nonholiday_sales = df[df['IsHoliday'] == False]['Weekly_Sales']
    print("Holiday weeks: count", holiday_sales.shape[0], "mean", holiday_sales.mean())
    print("Non-holiday weeks: count", nonholiday_sales.shape[0], "mean", nonholiday_sales.mean())
    # visual
    plt.figure(figsize=(8,5))
    sns.kdeplot(nonholiday_sales.dropna(), label='Non-holiday', fill=True)
    sns.kdeplot(holiday_sales.dropna(), label='Holiday', fill=True)
    plt.legend()
    plt.title("Weekly Sales: Holiday vs Non-holiday")
    plt.show()

# Decision guidance (printed; you can adjust to business rules)
print("""
Outlier treatment guidance:
- If outliers are data errors -> remove
- If outliers correspond to known holiday spikes -> keep as business important
- If extreme but plausible -> consider capping (winsorize)
""")

# Example: cap Weekly_Sales using IQR bounds instead of removing
if 'Weekly_Sales' in df.columns:
    w = pd.to_numeric(df['Weekly_Sales'], errors='coerce')
    l, u = iqr_outlier_bounds(w.dropna())
    df['Weekly_Sales_capped'] = w.clip(lower=l, upper=u)

# -----------------------------
# Task 4 — Univariate Analysis
# -----------------------------
# Weekly sales distribution
if 'Weekly_Sales' in df.columns:
    plt.figure(figsize=(10,5))
    sns.histplot(df['Weekly_Sales'].dropna(), kde=True, bins=50)
    plt.title("Weekly Sales distribution (all stores)")
    plt.show()

# Store type distribution
if 'Type' in df.columns:
    plt.figure(figsize=(6,4))
    sns.countplot(data=df, x='Type', order=df['Type'].value_counts().index)
    plt.title("Store Type Distribution")
    plt.show()

# Distributions for numeric features
for col in ['Temperature', 'Fuel_Price', 'CPI', 'Unemployment']:
    if col in df.columns:
        plt.figure(figsize=(8,3))
        sns.histplot(df[col].dropna(), kde=True)
        plt.title(f"Distribution of {col}")
        plt.show()

# Sales during holiday / non-holiday
if 'IsHoliday' in df.columns and 'Weekly_Sales' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(x='IsHoliday', y='Weekly_Sales', data=df)
    plt.title("Weekly Sales: Holiday vs Non-holiday (boxplot)")
    plt.show()

# Top 10 departments by average weekly sales
if 'Dept' in df.columns and 'Weekly_Sales' in df.columns:
    top10_dept = df.groupby('Dept')['Weekly_Sales'].mean().nlargest(10).reset_index()
    print("Top 10 departments by avg weekly sales:")
    display(top10_dept)
    plt.figure(figsize=(10,5))
    sns.barplot(data=top10_dept, x='Dept', y='Weekly_Sales')
    plt.title("Top 10 Depts by Avg Weekly Sales")
    plt.show()

# -----------------------------
# Task 5 — Bivariate Analysis
# -----------------------------
# Temperature vs Weekly Sales (scatter)
if 'Temperature' in df.columns and 'Weekly_Sales' in df.columns:
    plt.figure(figsize=(8,5))
    sns.scatterplot(data=df, x='Temperature', y='Weekly_Sales', alpha=0.4)
    plt.title("Temperature vs Weekly Sales")
    plt.show()

# Fuel Price vs Weekly Sales
if 'Fuel_Price' in df.columns and 'Weekly_Sales' in df.columns:
    plt.figure(figsize=(8,5))
    sns.scatterplot(data=df, x='Fuel_Price', y='Weekly_Sales', alpha=0.4)
    plt.title("Fuel Price vs Weekly Sales")
    plt.show()

# Weekly Sales vs Store Type
if 'Type' in df.columns and 'Weekly_Sales' in df.columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=df, x='Type', y='Weekly_Sales')
    plt.title("Weekly Sales by Store Type")
    plt.show()

# Sales vs Holiday Flag
if 'IsHoliday' in df.columns and 'Weekly_Sales' in df.columns:
    plt.figure(figsize=(6,4))
    sns.barplot(data=df, x='IsHoliday', y='Weekly_Sales')
    plt.title("Average Weekly Sales by Holiday Flag")
    plt.show()

# Compare top-performing store vs lowest-performing store
if 'Store' in df.columns and 'Weekly_Sales' in df.columns:
    store_sales = df.groupby('Store')['Weekly_Sales'].mean().sort_values(ascending=False)
    top_store, low_store = store_sales.index[0], store_sales.index[-1]
    print(f"Top store: {top_store}, Lowest store: {low_store}")
    # Plot both
    plt.figure(figsize=(12,4))
    sns.lineplot(data=df[df['Store']==top_store].sort_values('Date'), x='Date', y='Weekly_Sales', label=f"Store {top_store}")
    sns.lineplot(data=df[df['Store']==low_store].sort_values('Date'), x='Date', y='Weekly_Sales', label=f"Store {low_store}")
    plt.title("Top vs Lowest Store Weekly Sales Over Time")
    plt.legend()
    plt.show()

# -----------------------------
# Task 6 — Multivariate Analysis
# -----------------------------
# Correlation heatmap (numeric only)
numeric_df = df.select_dtypes(include=['number'])
plt.figure(figsize=(10,8))
sns.heatmap(numeric_df.corr(), annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Numeric Feature Correlations")
plt.show()

# Store-level analysis: use Type, Size, Weekly_Sales
if set(['Type','Size','Weekly_Sales']).issubset(set(df.columns)):
    store_summary = df.groupby(['Store','Type','Size'])['Weekly_Sales'].mean().reset_index()
    print("Store-level summary (sample):")
    display(store_summary.head())

# Weekly Sales vs Temperature vs Holiday_Flag (grouped summary)
if set(['Weekly_Sales','Temperature','IsHoliday']).issubset(set(df.columns)):
    grouped = df.groupby(['IsHoliday']).agg(
        avg_temp=('Temperature','mean'),
        avg_sales=('Weekly_Sales','mean'),
        count=('Weekly_Sales','count')
    ).reset_index()
    print("Grouped summary by IsHoliday:")
    display(grouped)

# Discount markdown influence: quick groupby
if set(markdown_cols).intersection(set(df.columns)):
    df['discount_effect'] = df[[c for c in markdown_cols if c in df.columns]].sum(axis=1)
    # analyze relation with sales across holidays
    pivot = df.groupby(['IsHoliday'])[['discount_effect','Weekly_Sales']].mean()
    print("Discount effect vs Weekly Sales by IsHoliday:")
    display(pivot)

# -----------------------------
# Task 7 — Time Series Analysis
# -----------------------------
if 'Date' in df.columns:
    df['year'] = df['Date'].dt.year
    df['month'] = df['Date'].dt.month
    df['week'] = df['Date'].dt.isocalendar().week

    # Total weekly sales over time (aggregated)
    weekly_total = df.groupby('Date')['Weekly_Sales'].sum().reset_index()
    plt.figure(figsize=(14,4))
    sns.lineplot(data=weekly_total, x='Date', y='Weekly_Sales')
    plt.title("Total Weekly Sales Over Time")
    plt.show()

    # Plot monthly sales trends for highest and lowest stores
    store_sales_total = df.groupby('Store')['Weekly_Sales'].sum().sort_values()
    highest_store = store_sales_total.index[-1]
    lowest_store = store_sales_total.index[0]
    print("Highest store:", highest_store, "Lowest store:", lowest_store)

    for s in [highest_store, lowest_store]:
        monthly = df[df['Store']==s].groupby('month')['Weekly_Sales'].sum().reset_index()
        plt.figure(figsize=(10,3))
        sns.lineplot(data=monthly, x='month', y='Weekly_Sales', marker='o')
        plt.title(f"Monthly Sales Trend - Store {s}")
        plt.xticks(range(1,13))
        plt.show()

    # Seasonal months: identify months with peak sales
    month_sales = df.groupby('month')['Weekly_Sales'].sum().sort_values(ascending=False)
    print("Months by total sales (top 6):")
    display(month_sales.head(6))

# -----------------------------
# Task 8 — Feature Engineering
# -----------------------------
# 8A: new columns
if 'Date' in df.columns:
    df['year'] = df['Date'].dt.year
    df['month'] = df['Date'].dt.month
    df['week'] = df['Date'].dt.isocalendar().week

# discount_effect (sum of markdowns)
existing_markdowns = [c for c in markdown_cols if c in df.columns]
if existing_markdowns:
    df['discount_effect'] = df[existing_markdowns].sum(axis=1)
else:
    df['discount_effect'] = 0.0

# is_peak_season
df['is_peak_season'] = df['month'].isin([11,12]) if 'month' in df.columns else False

# normalized_sales = Weekly_Sales / Size (if Size exists)
if 'Size' in df.columns and 'Weekly_Sales' in df.columns:
    df['normalized_sales'] = df['Weekly_Sales'] / df['Size'].replace({0: np.nan})

# 8B: filtering using created columns
# Peak-season transactions with weekly_sales > 50,000
if 'is_peak_season' in df.columns and 'Weekly_Sales' in df.columns:
    peak_big = df[(df['is_peak_season']==True) & (df['Weekly_Sales'] > 50000)]
    print("Peak-season transactions with Weekly_Sales > 50,000:", peak_big.shape[0])

# Stores whose normalized_sales in top 10%
if 'normalized_sales' in df.columns:
    thresh = df['normalized_sales'].quantile(0.9)
    top_norm = df[df['normalized_sales'] >= thresh]
    print("Stores/rows in top 10% normalized_sales:", top_norm.shape[0])

# Departments where discount_effect > median discount
if 'discount_effect' in df.columns:
    med_disc = df['discount_effect'].median()
    dept_high_disc = df.groupby('Dept')['discount_effect'].sum().sort_values(ascending=False).head(20)
    print("Departments by total discount_effect (sample):")
    display(dept_high_disc.head(10))

# Filter combined complex condition example
if set(['Temperature','Fuel_Price','Weekly_Sales','is_peak_season']).issubset(df.columns):
    filtered_complex = df[
        (df['Temperature'] < 40) &
        (df['Fuel_Price'] > 3.5) &
        (df['Weekly_Sales'].between(20000,60000)) &
        (df['is_peak_season'] == True)
    ]
    print("Filtered complex rows:", filtered_complex.shape[0])

# 8C: Grouping and business insights
# Monthly average sales per store
monthly_store = df.groupby(['Store','month'])['Weekly_Sales'].mean().reset_index().rename(columns={'Weekly_Sales':'avg_monthly_weekly_sales'})
print("Monthly average weekly sales per store (sample):")
display(monthly_store.head())

# Total discount_effect per department
if 'Dept' in df.columns:
    dept_discount = df.groupby('Dept')['discount_effect'].sum().reset_index().sort_values('discount_effect', ascending=False)
    print("Total discount_effect per Dept (top 10):")
    display(dept_discount.head(10))

# Department with highest normalized_sales (mean)
if 'normalized_sales' in df.columns and 'Dept' in df.columns:
    dept_norm = df.groupby('Dept')['normalized_sales'].mean().sort_values(ascending=False).reset_index()
    print("Department with highest normalized_sales (top 10):")
    display(dept_norm.head(10))

# revenue_potential = Weekly_Sales * 52 (annualized)
if 'Weekly_Sales' in df.columns:
    store_revenue = df.groupby('Store')['Weekly_Sales'].mean().mul(52).reset_index().rename(columns={'Weekly_Sales':'revenue_potential'})
    top10_revenue = store_revenue.sort_values('revenue_potential', ascending=False).head(10)
    print("Top 10 stores by revenue potential (annualized):")
    display(top10_revenue)

# -----------------------------
# Final: Save cleaned merged dataset (optional)
# -----------------------------
# df.to_csv("walmart_merged_cleaned.csv", index=False)
print("Pipeline complete. 'df' holds the cleaned merged DataFrame.")
```
